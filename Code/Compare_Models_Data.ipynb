```python
import pandas as pd
import matplotlib.pylab as plt
from matplotlib import ticker
```

# Comparing FWHMs

```python
FWHMs = pd.read_csv('FWHMs.csv',sep=',',index_col=[0])
FWHMs.Data = FWHMs.Data/1e3
FWHMs.Uncertainty = FWHMs.Uncertainty/1e3
FWHMs
```

```python
fig,(ax1,ax2) = plt.subplots(2,1,sharex=True)

names1 = ['B87']
data1 = [FWHMs[FWHMs.index == 11].Data.values[0]]
dy1 = [FWHMs[FWHMs.index == 11].Uncertainty.values[0]]

names2 = ['RTP','Observed']
data2 = [FWHMs[FWHMs.index == 1].Data.values[0],FWHMs[FWHMs.index == 2].Data.values[0]]
dy2 = [FWHMs[FWHMs.index == 1].Uncertainty.values[0],FWHMs[FWHMs.index == 2].Uncertainty.values[0]]

names3 = ['a','b','c']
data3 = [FWHMs[FWHMs.index == 5].Data.values[0],FWHMs[FWHMs.index == 6].Data.values[0],FWHMs[FWHMs.index == 7].Data.values[0]]
dy3 = [FWHMs[FWHMs.index == 5].Uncertainty.values[0],FWHMs[FWHMs.index == 6].Uncertainty.values[0],FWHMs[FWHMs.index == 7].Uncertainty.values[0]]

names4 = ['a','b','c']
data4 = [FWHMs[FWHMs.index == 8].Data.values[0],FWHMs[FWHMs.index == 9].Data.values[0],FWHMs[FWHMs.index == 10].Data.values[0]]
dy4 = [FWHMs[FWHMs.index == 8].Uncertainty.values[0],FWHMs[FWHMs.index == 9].Uncertainty.values[0],FWHMs[FWHMs.index == 10].Uncertainty.values[0]]

ax1.errorbar(names1,data1,dy1,fmt='x',label='B87',c='k')
ax1.errorbar(names2,data2,dy2,fmt='D',label='Processed',c='k')
ax1.errorbar(names3,data3,dy3,fmt='o',label='2.75D',c='k')
ax1.errorbar(names4,data4,dy4,fmt='s',label='2.5D',c='k')
# ax1.axhline(3.3*2,label='Expected',ls='--',c='k')
ax1.set_title('a)',loc='left')

names1 = ['B87']
data1 = [FWHMs[FWHMs.index == 12].Data.values[0]]
dy1 = [FWHMs[FWHMs.index == 12].Uncertainty.values[0]]

names2 = ['Observed']
data2 = [FWHMs[FWHMs.index == 0].Data.values[0]]
dy2 = [FWHMs[FWHMs.index == 0].Uncertainty.values[0]]

names3 = ['a','b']
data3 = [FWHMs[FWHMs.index == 3].Data.values[0],FWHMs[FWHMs.index == 4].Data.values[0]]
dy3 = [FWHMs[FWHMs.index == 3].Uncertainty.values[0],FWHMs[FWHMs.index == 4].Uncertainty.values[0]]

ax2.errorbar(names1,data1,dy1,fmt='x',label='B87',c='k')
ax2.errorbar(names2,data2,dy2,fmt='D',label='Processed',c='k')
ax2.errorbar(names3,data3,dy3,fmt='s',label='2.5D',c='k')
# ax2.axhline(10*2,label='Expected',ls='--',c='k')
ax2.set_title('b)',loc='left')

# ax1.legend(loc=2,bbox_to_anchor=(0.2,0.9))

fig.tight_layout(pad = 2)
fig.supylabel('FWHM [km]')
fig.show()
fig.savefig('FWHM_Comparison.pdf')
```

# Comparing SSR/ndf

```python
SSRs = pd.read_csv('Modelling/Gravity/SSR_SSRndf_model.csv',sep=',',index_col=[0])
SSRndf = SSRs[SSRs.DataType == 'SSR/ndf']
SSRndf
```

```python
fig,ax = plt.subplots()

names1 = ['a','b']
data1 = [SSRndf[SSRndf.Model == 'Grav1'].Data.values[0],SSRndf[SSRndf.Model == 'Grav2'].Data.values[0]]

names2 = ['a','b','c']
data2 = [SSRndf[SSRndf.Model == 'Mag1_2.5'].Data.values[0],SSRndf[SSRndf.Model == 'Mag2_2.5'].Data.values[0],SSRndf[SSRndf.Model == 'Mag3_2.5'].Data.values[0]]
data3 = [SSRndf[SSRndf.Model == 'Mag1_2.75'].Data.values[0],SSRndf[SSRndf.Model == 'Mag2_2.75'].Data.values[0],SSRndf[SSRndf.Model == 'Mag3_2.75'].Data.values[0]]

ax.scatter(names2,data2,marker='o',label='2.75D',c='r')
ax.scatter(names2,data3,marker='s',label='2.5D',c='r')
ax.scatter(names1,data1,marker='s',label='2.5D',c='k')
ax.yaxis.set_major_locator(ticker.MultipleLocator(0.005))
ax.set_xlabel('Model')

ax.set_ylabel(r'$\mathrm{SSR/n_{df}}$')
fig.set_size_inches(3,2)
fig.tight_layout()
fig.show()
fig.savefig('SSR_Comparison.pdf')
```
